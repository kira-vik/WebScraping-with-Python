```python
# import dependencies

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
books = [] # create empty books list

for i in range(1, 51):
     
     # set url of website to be scraped
     SCRAPE_URL = f'https://books.toscrape.com/catalogue/page-{i}.html'
     
     # send GET request to specified url
     response = requests.get(SCRAPE_URL)
     soup = BeautifulSoup(response.content, 'html.parser')
     soup = BeautifulSoup(soup.prettify(), 'html.parser')
     
     
     articles = soup.find_all('article', class_ = 'product_pod')

     # fetch data of interest; store data in the books list initialized earlier
     for article in articles:
          title = article.find('h3').find('a').attrs['title']
          rating = article.find('p')['class'][1]
          price = float(article.find('p', class_ = 'price_color').text.strip()[1:])
          
          books.append([title, price, rating]) # populate books list
```

```python
# checkout books list to confirm data retrieval successfully 
books
```

```python
# create pandas Dataframe; populate with books list
df = pd.DataFrame(books, columns=['Book Title', 'Price', 'Star Rating'])

df
```

```python
# export dataframe to csv
df.to_csv('BookStore.csv')
```
